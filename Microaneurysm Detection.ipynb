```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def detect_microaneurysms(retinal_image_path):
    def adjust_gamma(image, gamma=1.0):
        table = np.array([((i / 255.0) ** gamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
        return cv2.LUT(image, table)

    def extract_ma(image):
        r, g, b = cv2.split(image)
        comp = 255 - g
        clahe = cv2.createCLAHE(clipLimit=5.0, tileGridSize=(8, 8))
        histe = clahe.apply(comp)
        adjustImage = adjust_gamma(histe, gamma=3)
        comp = 255 - adjustImage
        J = adjust_gamma(comp, gamma=4)
        J = 255 - J
        J = adjust_gamma(J, gamma=4)

        K = np.ones((11, 11), np.float32)
        L = cv2.filter2D(J, -1, K)

        ret3, thresh2 = cv2.threshold(L, 125, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
        kernel2 = np.ones((9, 9), np.uint8)
        tophat = cv2.morphologyEx(thresh2, cv2.MORPH_TOPHAT, kernel2)
        kernel3 = np.ones((7, 7), np.uint8)
        opening = cv2.morphologyEx(tophat, cv2.MORPH_OPEN, kernel3)

        # Increase the size of microaneurysms using dilation
        kernel_dilation = np.ones((18, 18), np.uint8)  # Adjust the kernel size as needed
        dilated_ma = cv2.dilate(opening, kernel_dilation, iterations=1)

        return dilated_ma

    def detect_and_draw_boxes(original_image, binary_mask):
        contours, hierarchy = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        image_with_boxes = original_image.copy()
        for contour in contours:
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(image_with_boxes, (x, y), (x+w, y+h), (255, 0, 0), 12)

        return image_with_boxes, contours

    # Read the retinal image
    fundus = cv2.imread(retinal_image_path)

    # Extract microaneurysms
    microaneurysms = extract_ma(fundus)

    # Draw boxes around microaneurysms on the original image
    image_with_boxes, contours = detect_and_draw_boxes(fundus, microaneurysms)

    # Display the original image and microaneurysms with squares
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(fundus, cv2.COLOR_BGR2RGB))
    plt.set_cmap('inferno')
    plt.axis('off')
    plt.text(700, 2300, r'Original Image')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.text(700, 2300, f'Total Microaneurysms: {len(contours)}')
    plt.show()

    # Save the image with detected microaneurysms
    cv2.imwrite("microaneurysm_detected_image.png", image_with_boxes)

    # Return the count of microaneurysms
    return len(contours)

# Example usage:
retinal_image_path = "16_left.jpeg"
microaneurysm_count = detect_microaneurysms(retinal_image_path)

# Now, you can use microaneurysm_count as a separate variable
print(f'Total Microaneurysms in {retinal_image_path}: {microaneurysm_count}')
```
